```python
url = "http://127.0.0.1:5000/cpn/"
def get_data(num):
    return requests.get(url + str(num)).json()
y_10 = get_data(10)
x_10 = [i for i in range(1, 11)]
```

```python
x_10
```

```python
y_10
```

```python
fig = plt.figure(figsize = (10, 8))
plt.bar(x_10, y_10)
plt.xlabel('порядковый номер')
plt.ylabel('число последовательности ЦМЧ')
plt.title('Первые {} чисел последовательности ЦМЧ'.format(len(y_10)))
plt.show()
```
